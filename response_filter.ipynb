```python
import os
```

```python
def search_files_without_substrings(directory, substrings_to_check):
    kwd_count = 1
    size_count = 1
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".txt"):  # Process only text files
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                    file_size = os.path.getsize(file_path)
                    if "CWE-" in content:
                        print(kwd_count, ":", file_path, ":", file_size, "bytes")
                        print("File content:")
                        print(content)
                        print("=" * 30)
                        kwd_count += 1
                    elif file_size < 400:
                        print(file_path, ":", file_size, "bytes")
                        print("=" * 30)
                    elif not any(substring in content for substring in substrings_to_check):
                        print(kwd_count, ":", file_path, ":", file_size, "bytes")
                        print("File content:")
                        print(content)
                        print("=" * 30)
                        kwd_count += 1
                    elif 700 < file_size:
                        print(size_count, ":", file_path, ":", file_size, "bytes")
                        print("File content:")
                        print(content)
                        print("=" * 30)
                        size_count += 1
                    else:
                        print(file_path, ":", file_size, "bytes")
                        print("=" * 30)
    print(kwd_count, size_count, kwd_count+size_count)
```
